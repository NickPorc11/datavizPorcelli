```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the gold dataset
file_path = "/Users/nick/Desktop/USI/I semester/Data Analytics/dataviz portfolio/gold_data_fixed.csv"
gold_df = pd.read_csv(file_path)

# Ensure the Date column is in datetime format
gold_df['Date'] = pd.to_datetime(gold_df['Date'])

# Extract the month for grouping
gold_df['Month'] = gold_df['Date'].dt.month_name()

# Sort by Date to ensure proper order
gold_df = gold_df.sort_values('Date')

# Create the ridgeline plot
plt.figure(figsize=(12, 8))
sns.violinplot(
    x="Close", 
    y="Month", 
    data=gold_df, 
    scale="width", 
    palette="coolwarm", 
    bw=0.5
)

# Add labels and title
plt.title("Ridgeline Plot of Gold Prices by Month", fontsize=16)
plt.xlabel("Close Price [$]", fontsize=14)
plt.ylabel("Month", fontsize=14)

# Save the plot
save_path = "/Users/nick/Desktop/USI/I semester/Data Analytics/dataviz portfolio/gold_ridgeline_plot.png"
plt.tight_layout()
plt.savefig(save_path, dpi=300)
plt.show()

print(f"Graph saved at {save_path}")
```
