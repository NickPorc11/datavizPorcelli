```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = "/Users/nick/Desktop/USI/I semester/Data Analytics/dataviz portfolio/datasetetfbtctask6.xlsx"
df = pd.read_excel(file_path)

# Ensure the Date column is in datetime format and sort by date
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values('Date')

# Determine whether the ETF value indicates a positive or negative trend
df['Trend'] = df['Adj Close'].diff()  # Compute the change
df['Color'] = df['Trend'].apply(lambda x: 'green' if x > 0 else 'red')  # Assign color

# Plot the bar chart with an adjusted y-axis
plt.figure(figsize=(12, 8))
plt.bar(df['Date'], df['Adj Close'], color=df['Color'], width=0.8)

# Add labels, legend, and title
plt.title("BTC ETF Daily Trends (2025 to Today)", fontsize=16)
plt.xlabel("Date", fontsize=14)
plt.ylabel("Adjusted Close Value [$]", fontsize=14)
plt.xticks(rotation=45)

# Adjust the y-axis limits to make bars smaller
plt.ylim(0, df['Adj Close'].max() * 1.5)

# Add a legend with custom labels
red_patch = plt.Rectangle((0, 0), 1, 1, fc="red", edgecolor='none')
green_patch = plt.Rectangle((0, 0), 1, 1, fc="green", edgecolor='none')
plt.legend([green_patch, red_patch], ['Buy Vol. > Sell Vol.', 'Sell Vol. > Buy Vol.'], loc='upper left', fontsize=12)

# Save and show the plot
save_path = "/Users/nick/Desktop/USI/I semester/Data Analytics/dataviz portfolio/btc_etf_trends_adjusted.png"
plt.tight_layout()
plt.savefig(save_path, dpi=300)
plt.show()

print(f"Graph saved at {save_path}")
```
